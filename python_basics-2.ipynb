### Conditional Statement

```python
#if condition
age = 22
if age >= 18:
    print("Adult")
```

```python
#if- else condtion
age = 15
if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
# if- elif(grading)
marks= int(input("Enter your marks "))
if marks >= 90:
    print("A")
elif marks >= 80:
    print("B")
elif marks >= 70:
    print("C")
else:
    print("Fail")
```

```python
def state(nums):
    return min(nums), max(nums), sum(nums)/len(nums)
```

```python
low, high, avg = state([10,20,30,40])
```

```python
print(low,high,avg)
```
